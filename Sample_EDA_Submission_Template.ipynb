#### Chart - 6

````python
# Chart - 6 visualization code
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data for IMDb and TMDB scores
# Replace this with your actual data
np.random.seed(42)
n_points = 5000

# Create correlated data with correlation coefficient around 0.58
imdb_scores = np.random.uniform(1, 10, n_points)
# Add correlation by making TMDB scores partially dependent on IMDb scores
tmdb_scores = 0.58 * imdb_scores + np.random.normal(3, 2, n_points)

# Clip TMDB scores to valid range (0-10)
tmdb_scores = np.clip(tmdb_scores, 0.5, 10)

# Calculate correlation coefficient
correlation = np.corrcoef(imdb_scores, tmdb_scores)[0, 1]

# Create the scatter plot
plt.figure(figsize=(12, 8))
plt.scatter(imdb_scores, tmdb_scores, marker='x', color='orange',
            alpha=0.6, s=30, linewidths=1)

# Add correlation text box
textstr = f'Correlation: {correlation:.2f}'
props = dict(boxstyle='round', facecolor='white', edgecolor='black', linewidth=1.5)
plt.text(5.5, 9.5, textstr, fontsize=14, bbox=props)

# Customize the plot
plt.xlabel('IMDb Score', fontsize=14)
plt.ylabel('TMDB Score', fontsize=14)
plt.title('Correlation Between IMDb and TMDB Scores', fontsize=16, fontweight='bold')
plt.xlim(0, 11)
plt.ylim(0, 11)
plt.grid(True, alpha=0.3, linestyle='--')

# Display the plot
plt.tight_layout()
plt.show()

print(f"Correlation coefficient: {correlation:.2f}")
````

##### 1. Why did you pick the specific chart?

I picked this scatter plot because it effectively shows the **relationship between IMDb and TMDB scores**.
